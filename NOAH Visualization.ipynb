```python
import pandas as pd
import folium
import numpy as np
```

```python
df = pd.read_csv('NOAH_Query.csv')
df = df.drop_duplicates(subset=['parcel_number'])
df = df[df['year_built'] < 1990]
df['popup'] = df.address + '\n' + df.owner + '\nYear Built: ' + df.year_built.astype(str) +'\nAuditor Value: ' + df.market_total_value_2020.astype(str)
df.head(5)
```

```python
print('There are ' + str(len(df)) + ' 20+ apartment buildings in Cincinnati'
```

# 40+ Unit Buildings Mapped

```python
# Map consisting of all the 40+ apartment complexes built before 1990
forty_df = df[df.property_class == 403]
m = folium.Map(location=[39.158, -84.51], zoom_start=12)

for x in range(len(forty_df)):
    coords = df.coordinates.iloc[x]
    lat = coords.split(',')[1]
    lat = float(lat.split(')')[0])
    long = coords.split(',')[0]
    long = float(long.split('(')[1])
    folium.Marker([lat, long], popup=df.popup.iloc[x]).add_to(m)

m
```
