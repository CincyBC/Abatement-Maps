```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('dillingham_results.csv')
df['registered'] = df.registered.str.replace(",", '')
df['registered'] = df.registered.astype(int)
df['precincts'] = df.registered.astype(str)
df['pct_registered'] = round(df.dillingham / df.registered, 4) * 100
df['pct_cast'] = round(df.dillingham / df.cast, 4) * 100
df.head(50)
```

```python
m = folium.Map(location=[39.135, -84.51], zoom_start=12, tiles='Stamen Toner')

folium.Choropleth(
    geo_data='Hamilton_County_Voter_Precincts.geojson',
    name='Dillingham Results',
    data=df,
    columns=['precincts', 'pct_cast'],
    key_on='feature.properties.PRECINCT',
    fill_color='OrRd',
    bins=[0, 10, 20, 30, 40, 60],
    legend_name='Percentage of Cast Ballots (%)'
).add_to(m)

folium.LayerControl().add_to(m)

m
```
